## Mais sobre plots

Cores: https://matplotlib.org/stable/gallery/color/named_colors.html

Linestyles: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

```python
fig, axis=plt.subplots()
x=np.linspace(0,1,1000)
y1=x**1
y2=x**2
y3=x**3
y4=x**4
Y2=x**(1/2)
Y3=x**(1/3)
Y4=x**(1/4)
axis.plot(x,y1,label=r'$x$',alpha=0.1)
axis.plot(x,y2,label=r'$x^2$',linewidth=3)
axis.plot(x,y3,label=r'$x^3',color='r')
axis.plot(x,y4,label=r'$x^4',linestyle='dashed')
axis.plot(x,Y2,label=r'$\sqrt{x}$',linestyle='dashdot')
axis.plot(x,Y3,label=r'$\sqrt[3]{x}$',linestyle='dotted')
axis.plot(x,Y4,label=r'$\sqrt[4]{x}$',color='saddlebrown')
axis.legend(loc='upper left')
fig.show()
```

## Pontos com margem de erro (errobar)

Em dados experimentais sempre há alguma margem de erro/incerteza do experimento, caso 
elas sejam grandes é interessantes representa-las em seu gráfico. Vamos usar a biblioteca [uncertainties](https://readthedocs.org/projects/uncertainties-python-package/downloads/pdf/latest/) para propagar os erros, a propagação é linear, para incertezas pequenas e funções que tem comportamento localmente linear funciona muito bem, para casos mais avançados é recomendado algo como um monte carlo que é fornecido pela biblioteca [mcerp](https://github.com/tisimst/mcerp)

(**Para os estudantes da USP**)
A biblioteca
[LabIFSC](https://github.com/gjvnq/LabIFSC) foi criada por estudantes da USP de São Carlos que foi pensada especialmente para o Laborátorio de Física, a propagação de erro e as equivalências são identicas ao usado na matéria. 

### Exemplo simples

```python
from uncertainties import ufloat
x=ufloat(2,0.2) #2+-0.2
y=ufloat(16,0.1) #16+-
print(f"Valor nominal {x.nominal_value}, incerteza {x.std_dev}")
print("-----")
print(f"Multipicação {x*y}")
print(f"Adição {x+y}")
print(f"Divisão {x/y}")
print("-----")
from uncertainties.umath import *
angulo=ufloat(1,0.1)
print(f"sin {sin(angulo)}")
print(f"cos {cos(angulo)}")
valor=ufloat(4,0.2)
print(f"exp {exp(valor)}")
print(f"log {log(valor)}")
```

### Erro constante

```python
x=[1,2,3,4]
x_err=0.1
y=[3.1,4.9,7.2,8.8]
y_err=0.2
figure,axis=plt.subplots()
axis.errorbar(x,y,x_err,y_err, fmt='o')
```

### Erro Variavel

```python
fig,axis=plt.subplots()

from uncertainties.unumpy import uarray, nominal_values,std_devs

valores = [2.3,3.7,5.6,7.8]
incerteza=0.2
dados = uarray(valores, incerteza)
print(dados)
dados=dados**2
print(dados)
y_nominais=nominal_values(dados)
y_erros=std_devs(dados)
axis.errorbar([2,3,4,5],y_nominais,yerr=y_erros,fmt='o')
#fig.show()
```

## Erro na curva téorica

Muitas vezes uma curva téorica é colocada em comparação a dados experimentos, porém, 
tanto a curva téorica quanto os dados possuem incertezas, como uma curva possui 
um número infinito de pontos faz mais sentido pensar em uma curva máxima e uma mínima


<img width=400 src="https://raw.githubusercontent.com/viniciusdutra314/CanalSingularidade/main/GraficosCientificos/curva_helmhotz.jpg" alt="Logo matplotlib"> 


```python
#vamos considerar uma queda livre y=(g/2)*t**2
#temos incerteza no g, consequemente temos sobre (y(t)) a curva toda
g=ufloat(9.5,0.5)
t=np.linspace(1,10,1000)
y=(g/2)*(t**2)

y_nominal=nominal_values(y)
y_min=nominal_values(y) -2*std_devs(y)
y_max=nominal_values(y) +2*std_devs(y)

figure, axis=plt.subplots()
axis.fill_between(t,y_min,y_max,alpha=0.3)
axis.plot(t,y_nominal,color='r')
figure.show()
```
