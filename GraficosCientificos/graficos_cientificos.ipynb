## 🔴 plt vs figure (importante)

No caso de figuras com somente um axis, é possível não criar uma figure e gerar o gráfico diretamento com plt (exemplo plt.scatter), o problema disso é que as sintaxes são levemente diferentes com e sem figuras, caso você queira adicionar mais axis para a sua figure será necessário mudar todo código, nesse notebook e em gráficos complexos eu recomendo sempre criar uma figure

| figure    | plt |
| -------- | ------- |
| set_xlabel  | xlabel    |
| set_xscale | xscale     |
| set_title    | title   |

```python
import pandas as pd
link='https://raw.githubusercontent.com/viniciusdutra314/CanalSingularidade/main/GraficosCientificos/osciloscopio.csv'
dataframe=pd.read_csv(link)
fig, axis=plt.subplots()
print(dataframe)
#No brasil usamos a virgula ao invés de ponto (exemplo: 0,5 e não 0.5)
#Isso geralmente faz com que o pandas importe os dados como texto e não como numeros
tempo=np.array([float(tempo.replace(",",".",)) for tempo in dataframe['Time']])
voltagem=np.array([float(voltagem.replace(",",".",))for voltagem in dataframe['Volt']])
axis.scatter(tempo,voltagem,s=5.5)
fig.show()
```

## Curva matemática (plots)

É feito uma interpolação linear entre os pontos fornecidos, é possível fazer
ela sem numpy, porém os arrays do numpy são mais rápidos
e simples de se manipular em termos matemáticos do que listas nativas do Python

!pip install numpy

```python
import numpy as np
x=np.linspace(-2,2,100) #explicar o que é o x
y=x**2
fig, axis=plt.subplots()
axis.plot(x,y,label="Muitos pontos (100)")
x_poucos=np.linspace(-2,2,10)
y_poucos=x_poucos**2+1
axis.plot(x_poucos,y_poucos,label="Poucos pontos (10)")
axis.legend(loc='upper center')
fig.show()
```

## Mais sobre plots

Cores: https://matplotlib.org/stable/gallery/color/named_colors.html

Linestyles: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

```python
fig, axis=plt.subplots()
x=np.linspace(0,1,1000)
y1=x**1
y2=x**2
y3=x**3
y4=x**4
Y2=x**(1/2)
Y3=x**(1/3)
Y4=x**(1/4)
axis.plot(x,y1,label=r'$x$',alpha=0.1)
axis.plot(x,y2,label=r'$x^2$',linewidth=3)
axis.plot(x,y3,label=r'$x^3$',color='red')
axis.plot(x,y4,label=r'$x^4$',linestyle='dashed')
axis.plot(x,Y2,label=r'$\sqrt{x}$',linestyle='dashdot')
axis.plot(x,Y3,label=r'$\sqrt[3]{x}$',linestyle='dotted')
axis.plot(x,Y4,label=r'$\sqrt[4]{x}$',color='saddlebrown')
axis.legend(loc='upper left')
fig.show()
```

## Plot + Scatter real

```python
import numpy as np
##dados (scatter)
f=np.array([1.20,2.34,3.42,4.60,5.72,6.78,7.81,8.96,9.75,10.49,11.22]) #kHz
n=np.linspace(1,11,11) #harmonicos
##melhor reta (Plot)
reta=np.polyfit(n,f,deg=1)
y=reta[0]*n+reta[1]
fig , axis=plt.subplots()
axis.scatter(n,f,label="Tubo Fechado $L=0.15m$")
axis.plot(n,y,label="Melhor reta", color='g')
axis.set_title("Harmônicos tubo fechado")
axis.set_ylabel(r"Frequências $f_n$ (kHz)")
axis.set_xlabel("$nº$ Harmônico")
axis.legend()
fig.show()
```

## Pontos com margem de erro (errobar)

Em dados experimentais sempre há alguma margem de erro/incerteza do experimento, caso
elas sejam grandes é interessantes representa-las em seu gráfico. A biblioteca
[LabIFSC](https://github.com/brenopelegrin/LabIFSC) fornece métodos fáceis de calcular incertezas, o link é um fork da biblioteca,
ele facilita as operações com arrays que veremos em seguida

### Erro constante

```python
x=[1,2,3,4]
x_err=0.2
y=[3.1,4.9,7.2,8.8]
y_err=0.1
fig,axis=plt.subplots()
axis.errorbar(x,y,x_err,y_err, fmt='o')
fig.show()
```

### Erro Variavel

```python
import LabIFSC as lab
import numpy as np
fig,axis=plt.subplots()
x=[2,3,4,5]
dados=np.array([2.3,3.7,5.6,7.8])
y=lab.M(dados,incerteza=0.2) #precisamos do quadrado dessas medidas
medidas_quadradas=(y**2)
y_nominais=[j.nominal for j in medidas_quadradas]
y_erros=[j.incerteza for j in medidas_quadradas]
axis.errorbar(x,y_nominais,yerr=y_erros,fmt='o')
fig.show()
```

## Erro na curva téorica

Muitas vezes uma curva téorica é colocada em comparação a dados experimentos, porém,
tanto a curva téorica quanto os dados possuem incertezas, como uma curva possui
um número infinito de pontos faz mais sentido pensar em uma curva máxima e uma mínima


<img width=400 src="https://raw.githubusercontent.com/viniciusdutra314/CanalSingularidade/main/GraficosCientificos/curva_helmhotz.jpg" alt="Logo matplotlib">


```python
#vamos considerar uma queda livre y=(g/2)*t^2
#temos incerteza no g, consequemente temos sobre (y(t)) a curva toda
g=lab.Medida((9.5,0.4))
t=np.linspace(1,10,100)
y=(lab.M(t)**2)*(g/2)

y_nominal=[j.nominal for j in y]
y_min=[j.nominal - 2*j.incerteza for j in y]
y_max=[j.nominal + 2*j.incerteza for j in y]
figure, axis=plt.subplots()
axis.fill_between(t,y_min,y_max,alpha=0.3)
axis.plot(t,y_nominal,color='r')
figure.show()
```
