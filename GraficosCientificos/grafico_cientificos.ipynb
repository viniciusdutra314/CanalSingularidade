## Show / Savefig

Dependendo de onde você está executando seu código (Jupyter-notebook, google colab) o gráfico

pode aparecer automaticamente ou não, para garantir que apareça temos os métodos .show e savefig

```python
fig,axis=plt.subplots(figsize=(10,5))
x=[1,2,3,4]
y=[1,4,9,16]
axis.plot(x,y)
fig.savefig('figura_salva.png',dpi=400)
fig.savefig('figura_salva.pdf',dpi=400)
fig.savefig('figura_salva.svg',dpi=400)
#Primeiro savefig depois show, a ordem inversa
#vai salvar uma imagem em branco
fig.show()
```

### Dados reais (.csv)

Geralmente vão estar guardados em grandes .csv ou .json. O exemplo abaixo é um 
.csv  que podemos acessar usando a biblioteca pandas

!pip install pandas

```python
import pandas as pd
dataframe=pd.read_csv('osciloscopio.csv')
fig, axis=plt.subplots()
print(dataframe)
#No brasil usamos a virgula ao invés de ponto (exemplo: 0,5 e não 0.5)
#Isso geralmente faz com que o pandas importe os dados como texto e não como numeros
tempo=np.array([float(tempo.replace(",",".",)) for tempo in dataframe['Time']])
voltagem=np.array([float(voltagem.replace(",",".",))for voltagem in dataframe['Volt']])
axis.scatter(tempo-min(tempo),voltagem,s=5.5)
fig.show()
```

## Mais sobre plots

Cores: https://matplotlib.org/stable/gallery/color/named_colors.html

Linestyles: https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

```python
fig, axis=plt.subplots()
x=np.linspace(0,1,1000)
y1=x**1
y2=x**2
y3=x**3
y4=x**4
Y2=x**(1/2)
Y3=x**(1/3)
Y4=x**(1/4)
axis.plot(x,y1,label=r'$x$',alpha=0.1)
axis.plot(x,y2,label=r'$x^2$',linewidth=3)
axis.plot(x,y3,label=r'$x^3',color='r')
axis.plot(x,y4,label=r'$x^4',linestyle='dashed')
axis.plot(x,Y2,label=r'$\sqrt{x}$',linestyle='dashdot')
axis.plot(x,Y3,label=r'$\sqrt[3]{x}$',linestyle='dotted')
axis.plot(x,Y4,label=r'$\sqrt[4]{x}$',color='saddlebrown')
axis.legend(loc='upper left')
fig.show()
```

## Plot + Scatter real 

```python
import numpy as np
##dados (scatter)
f=np.array([1.20,2.34,3.42,4.60,5.72,6.78,7.81,8.96,9.75,10.49,11.22]) #kHz
n=np.linspace(1,11,11) #harmonicos
##melhor reta (Plot)
reta=np.polyfit(f,n,deg=1)
y=reta[0]*(n+1)+reta[1]
fig , axis=plt.subplots()
axis.scatter(n,f,label="Tubo Fechado $L=0.15m$")
axis.plot(n,y,label="Melhor reta", color='g')
#axis.set_title("Harmônicos tubo fechado")
#axis.set_ylabel(f"Frequências $f_n$ (kHz)")
#axis.set_xlabel("$nº$ Harmônico")
#axis.legend()
fig.show()
```

## Linhas verticais e horizontais

```python
x=np.array(range(10))
y=np.array([33,64,31,49,101,51,76,13,47,9])
fig , axis =plt.subplots()
axis.scatter(x,y)
media=np.mean(y)
#hline=horizontal line
#vline=vertical line
axis.axhline(media,color='g',label=f'Valor médio {media}')
axis.axvline(4,label='Exemplo de linha vertical',alpha=0.4,color='r')
axis.legend()
fig.show()
```
