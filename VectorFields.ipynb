## Referências:
[Stream lines](https://docs.manim.community/en/stable/reference/manim.mobject.vector_field.StreamLines.html#manim.mobject.vector_field.StreamLines)
[Arrow Vector Field](https://docs.manim.community/en/stable/reference/manim.mobject.vector_field.ArrowVectorField.html#manim.mobject.vector_field.ArrowVectorField)

```python

%%manim -ql BasicUsage

class BasicUsage(Scene):
    def construct(self):
        func = lambda pos: np.sin(pos[1]) * RIGHT + np.cos(pos[0]) * UP #define a vector field
        scale = lambda x: x/3 #how to set the size of the vectors
        vector_field = ArrowVectorField(func,length_func=scale)
        self.add(vector_field)
        self.play(StreamLines(func,opacity=0.3,stroke_width=0.01,color=RED).create()) #create streamlines
        self.wait()
```
