```python
%%manim -qk AlphaError
x=np.array([13.083333333333334, 19.033333333333335, 7.0, 17.083333333333332, 9.116666666666667, 5.15, 16.116666666666667, 17.15, 12.133333333333333, 13.1, 9.15, 5.05, 14.15, 18.15, 13.066666666666666, 9.083333333333334, 18.066666666666666, 10.0, 8.15, 6.133333333333334, 10.116666666666667, 16.1, 7.133333333333334, 14.066666666666666])
y=np.array([7.016666666666667, 10.0, 3.1333333333333333, 9.083333333333334, 5.133333333333334, 3.15, 9.116666666666667, 9.116666666666667, 7.15, 7.066666666666666, 5.116666666666666, 3.0166666666666666, 7.0, 9.066666666666666, 6.15, 5.1, 9.033333333333333, 5.133333333333334, 4.116666666666666, 3.1166666666666667, 5.15, 6.0, 4.083333333333333, 7.05])
def erro(alpha,beta):
  global x,y
  erro=0
  for j in range(len(x)):
    erro+=(y[j]-x[j]*alpha-beta)**2
  return erro
alpha=np.linspace(0,1,1000)
error=[]
for j in alpha:
  error.append(erro(j,0))
print(error)
print(np.polyfit(x,y,1))
class AlphaError(Scene):
  def construct(self):
    axes = Axes(x_range=[0, 1, 0.1],y_range=[0, 4, 1],axis_config={"color": BLUE},x_axis_config={
                "numbers_to_include": np.arange(0, 1, 0.1), 
                "numbers_with_elongated_ticks": np.arange(0, 1, 0.1)
             }, y_axis_config={"scaling": LogBase(custom_labels=True),"numbers_to_include": np.arange(0, 11, 2),"numbers_with_elongated_ticks": np.arange(0, 11, 2)})
    self.play(Write(axes),run_time=4)   
    labels =axes.get_axis_labels(x_label=r'\alpha',y_label='Erro').set_color(GREEN)
    self.play(Write(labels))
    graph=axes.plot_line_graph(alpha,error, vertex_dot_radius=0,line_color=RED)
    self.play(FadeIn(graph),run_time=5)
    self.wait()
```
