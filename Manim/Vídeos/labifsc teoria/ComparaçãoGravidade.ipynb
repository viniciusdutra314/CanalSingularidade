```python
from manim import *

config.media_width = "75%"
config.verbosity = "WARNING"
```

```python

%%manim -qk CompareNumbers


class CompareNumbers(Scene):
    def construct(self):
      numbers = NumberLine(
            x_range=[0, 10],include_numbers=True)
      self.play(FadeIn(numbers),run_time=2)
      self.wait(2)
      numbers_zoom=Axes(x_range=[9.5, 10.1,0.5],x_length=9*0.9 ,x_axis_config={"numbers_to_include": np.arange(9.5, 10.1, 0.1)},tips=False,y_length=0.001)
      numbers_zoom.shift(RIGHT)
      g1=NumberLine(x_range=[9.64,9.8,0.16],include_numbers=True,length=2.4*0.9,color=BLUE)
      texto_zoom=Text("[9.5 , 10.1]")
      self.play(Transform(numbers,texto_zoom),run_time=2)
      self.wait(2)
      self.play(Transform(numbers,numbers_zoom))
      g0=NumberLine(x_range=[9.79,9.81,0.02],length=0.3*0.9,color=RED)
      g0.shift(DOWN)
      self.play(Create(g0),run_time=2)
      gravidadereal=Text("Gravidade =[9.79,9.81]",color=RED,font_size=32)
      gravidadereal.shift(DOWN*2)
      self.play(Create(gravidadereal),run_time=2)
      g1.move_to(numbers_zoom.coords_to_point(9.72,0))
      g1.shift(UP)
      g2=NumberLine(x_range=[9.3,9.4,0.1],include_numbers=True,length=1.5*0.9,color=GREEN)
      g2.move_to(numbers_zoom.coords_to_point(9.35,0))
      g2.shift(2*UP)
      self.wait(2)
      self.play(Create(g1),run_time=2)
      self.wait(2)
      self.play(Create(g2),run_time=2)
      errado=Text("Errado",color=GREEN,font_size=20)
      errado.move_to(numbers_zoom.coords_to_point(9.35,0))
      errado.shift(UP*2.5)
      self.play(Create(errado))
      self.wait(2)
```
