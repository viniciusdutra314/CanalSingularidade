```python
from manim import *
config.media_width = "75%"
config.verbosity = "WARNING"
```

```python
%%manim -ql Text

class Text(Scene):
  def construct(self):
    text = MathTex(r"\sum m_i \vec{v_i}=0",color=BLUE,font_size=120,tex_template=TexFontTemplates.helvetica_fourier_it)
    box = SurroundingRectangle(text, color=GREEN,fill_color=GREEN,fill_opacity=0.3,buff=MED_LARGE_BUFF)
    box.z_index=-1
    self.play(DrawBorderThenFill(box))
    self.play(Write(text))
    self.wait(2)
    self.play(Unwrite(text))
```
